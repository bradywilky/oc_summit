```python
# If requests isn't installed in your notebook environment, uncomment:
# %pip install requests

import os
import json
import requests

# ----------------------------
# SET THESE TWO VALUES
# ----------------------------


API_BASE = "https://discord.com/api/v10"
HEADERS = {
    "Authorization": f"Bot {BOT_TOKEN}",
    "Content-Type": "application/json",
}

def discord_request(method, endpoint, **kwargs):
    url = f"{API_BASE}{endpoint}"
    resp = requests.request(method, url, headers=HEADERS, timeout=30, **kwargs)

    print(f"{method} {url}")
    print("Status:", resp.status_code)

    try:
        data = resp.json()
    except ValueError:
        data = resp.text

    print("Response:")
    print(json.dumps(data, indent=2) if isinstance(data, dict) else data)
    print("-" * 60)

    resp.raise_for_status()
    return data

# 1) Verify the token / identify the bot user
me = discord_request("GET", "/users/@me")

# 2) Send a test message to your channel
payload = {
    "content": "Hello from my Jupyter notebook test bot 🤖"
}

message = discord_request(
    "POST",
    f"/channels/{CHANNEL_ID}/messages",
    json=payload
)

print("Done. Message ID:", message["id"])
```
