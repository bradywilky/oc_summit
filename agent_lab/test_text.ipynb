```python
import time
from twilio.rest import Client


client = Client(ACCOUNT_SID, AUTH_TOKEN)

message = client.messages.create(
    body="Test SMS from Jupyter",
    from_=TWILIO_NUMBER,
    to=YOUR_NUMBER
)

print("Initial SID:", message.sid)
print("Initial status:", message.status)

time.sleep(5)

updated = client.messages(message.sid).fetch()

print("Updated status:", updated.status)
print("Error code:", updated.error_code)
print("Error message:", updated.error_message)
print("To:", updated.to)
print("From:", updated.from_)
```
